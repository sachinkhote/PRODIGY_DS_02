```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = pd.read_csv("test.csv")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)
titanic.dropna(subset=['Embarked'], inplace=True)

# Verify that missing values have been handled
titanic.isnull().sum()
```

```python
# Explore the distribution of passengers in each class
sns.countplot(x='Pclass', data=titanic)
plt.title('Passenger Count in Each Class')
plt.show()
```

```python

# Explore the distribution of passengers based on gender
sns.countplot(x='Sex', data=titanic)
plt.title('Passenger Count Based on Gender')
plt.show()
```

```python

# Explore the distribution of passengers based on age
sns.histplot(x='Age', data=titanic, bins=30, kde=True)
plt.title('Passenger Age Distribution')
plt.show()
```

```python
# Explore the relationship between class and age
sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.title('Relationship between Class and Age')
plt.show()
```

```python
# Explore the correlation between Siblings/Spouses (SibSp) and Parents/Children (Parch)
sns.heatmap(titanic[['SibSp', 'Parch']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between SibSp and Parch')
plt.show()
```

```python

# Explore the distribution of fares
sns.histplot(x='Fare', data=titanic, bins=30, kde=True)
plt.title('Distribution of Fares')
plt.show()
```

```python
# Explore the distribution of passengers based on embarkation point
sns.countplot(x='Embarked', data=titanic)
plt.title('Passenger Count Based on Embarked')
plt.show()
```
